```python
original_data = 'model_6_80/example.cpp'
data_paths = 'model_6_80/example.vw'
output = open(data_path, 'w')
with open(original_data, 'r') as file:
            data = file.read().replace('\n', ' ') # Заменяем всяческие отступы на пробелы
            data = data.replace('\t', ' ')
            data = data.replace('|', ' ')
            data = data.replace('.', ' ')
            data = data.replace('    ', ' ').lower()
            data = data.replace(':', ' dotdot')
            output.write('{} |text {} |class_id {}\n'.format(1, data, 'cpp')) 
output.close()
```
